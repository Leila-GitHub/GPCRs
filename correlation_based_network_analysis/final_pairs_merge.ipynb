```python
def merge_final_pairs(Lig_name):

    all_files = sorted(glob.glob(Lig_name + '_only_final_pairs_*.csv'))
    all_files

    data_frames = []
    for file_ in all_files:
        df = pd.read_csv(file_,index_col=None, header=0)
        data_frames.append(df)
    
    merged = reduce(lambda  left,right: pd.merge(left,right,on=["index_1","index_2"],
                       how='outer'), data_frames).fillna('not_common')

    common_merged = merged[~merged.eq('not_common').any(1)] 
    pd.DataFrame.to_csv(common_merged, Lig_name + '_merged_final_pairs.csv', sep=',', index=False)

    return common_merged
```
